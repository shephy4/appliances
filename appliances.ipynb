```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data = data.drop(['date', 'lights'], axis = 1)
data
```

```python
normalised_data = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
normalised_data
```

```python
target_variable = normalised_data['Appliances']
target_variable = pd.DataFrame(target_variable)
target_variable
```

```python
data_2 = data.drop('Appliances', axis = 1)
data_2
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data_2, target_variable, test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
pred_y = linear_model.predict(x_test)
pred_y = pd.DataFrame(pred_y, columns = ['predictions'])
pred_y
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, pred_y)
round(mae, 2) 
```

```python
#R-Squared.py
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, pred_y)
round(r2_score, 2)
```

```python
#Residual Sum of Squares.py
import numpy as np
rss = np.sum(np.square(y_test - pred_y))
round(rss, 2)
```

```python
#Root Mean Square Error.py
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, pred_y))
round(rmse, 3)
```

```python
def get_weights_df(model, feat, col_name):
    
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(x_train, y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)

lasso_reg.fit(x_train, y_train)
```

```python
linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')
```

```python
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```
